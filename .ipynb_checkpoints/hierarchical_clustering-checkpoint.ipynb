```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

from adni_util import get_cohort_data
```

```python
CN_COHORT_DIR = '/home/hice1/khom9/scratch/ADNI_Data/Hierarchical-Clustering/CN_Cohort'
MCI_COHORT_DIR = '/home/hice1/khom9/scratch/ADNI_Data/Hierarchical-Clustering/MCI_Cohort'
AD_COHORT_DIR = '/home/hice1/khom9/scratch/ADNI_Data/Hierarchical-Clustering/AD_Cohort'
```

```python
ad_imgs, ad_patient_info = get_cohort_data(AD_COHORT_DIR, normalize=False)
```

```python
df = pd.DataFrame(ad_patient_info)
df.columns = ['patient_id', 'acq_date']
# df.groupby('patient_id').apply(lambda x: x['acq_date'].sort_values().tolist())
df
```

```python
ad_imgs = np.stack(ad_imgs)
```

```python
ad_imgs.shape
```

```python
ad_imgs_norm = ad_imgs / np.expand_dims(ad_imgs.mean(axis=(1,2,3)), axis=(1,2,3))
```

```python
# model = AgglomerativeClustering(compute_full_tree=True)
model = AgglomerativeClustering(n_clusters=3)
model.fit(ad_imgs_norm.reshape(ad_imgs_norm.shape[0], -1))
```

```python
pd.Series(model.labels_).value_counts() / model.labels_.shape
```

```python
df['label'] = model.labels_
df
```

```python
df_first_visit = df.sort_values(by='acq_date').drop_duplicates(subset='patient_id', keep='first')
df_first_visit['label'].value_counts() / len(df_first_visit)
```

```python
linked_matrix = linkage(ad_imgs_norm.reshape(ad_imgs_norm.shape[0], -1), method='ward')
```

```python
plt.figure(figsize=(10, 7))
dendrogram(linked_matrix,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data points')
plt.ylabel('Distance')
plt.show()
```
