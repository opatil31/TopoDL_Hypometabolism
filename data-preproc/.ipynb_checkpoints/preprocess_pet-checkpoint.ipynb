```python
# mm offset calculated from SPM8 that is needed to re-set the origin to the ACM
OFFSET = np.array([118.5, 120, -72])
```

```python
def get_nii_affine(nii_path):
    '''
    Given the path of the NIFTI file, return only the affine matrix
    '''
    img = nib.load(nii_path)
    return img.affine


def add_to_header(nii_path, data):
    '''Add extra metadata to the header of the NIFTI image. Should be a dict or dict-like'''
    img = nib.load(nii_path)
    for k, v in data.items():
        img.header[k] = v
    # print(img.extra)
    nib.save(img, nii_path)

def get_dcm_paths(cohort_dir):    
    for patient_id in os.listdir(cohort_dir):
        patient_id_path = os.path.join(cohort_dir,patient_id)
        for img_preproc_type in os.listdir(patient_id_path):
            img_preproc_path = os.path.join(patient_id_path, img_preproc_type)
            for visit_date in os.listdir(img_preproc_path):
                visit_date_path = os.path.join(img_preproc_path, visit_date)
                for img_id in os.listdir(visit_date_path):
                    dcm_path = os.path.join(visit_date_path, img_id)
                    yield dcm_path
```

```python
def convert_all_dicom(cohort_dir, nii_dir):
    '''
    Convert all the images in a cohort from individual DICOM slices into NIFTI files (.nii)

    cohort_dir: Path to ADNI cohort which contains directories labeled by patient ID.
    nii_dir: Output directory to write to.
    '''
    if not os.path.exists(cohort_dir):
        raise ValueError(f'Cohort directory does not exist: "{cohort_dir}"')
    if not os.path.exists(nii_dir):
        os.mkdir(nii_dir)
        print(f'Created NII directory: "{nii_dir}"')
    
    eng = matlab.engine.start_matlab()
    eng.cd(os.getcwd())
    for path in tqdm(list(get_dcm_paths(cohort_dir))):
        img_id = os.path.basename(path)[1:]
        out = eng.dicom_to_nii(path, nii_dir, nargout=1)
        print(out)
        out_path = out['files'][0]
        out_dir = os.path.dirname(out_path)
        new_name = f'{img_id}-{os.path.basename(out_path)}'
        os.rename(out_path, os.path.join(out_dir, new_name))

    eng.quit()
```

```python
# Check that all affine matrices of the scans are equal

affines = []
for nii_name in os.listdir(MCI_COHORT_NII_DIR):
    affine = get_nii_affine(os.path.join(MCI_COHORT_NII_DIR, nii_name))
    affines.append(affine)
# True means all affine matrices are equal
print(np.all([np.allclose(affines[0], affines[i]) for i in range(len(affines))]))
```

```python
# Reset the origins for all images to the center of the image (near ACM). For preprocessed ADNI images, this is (80, 80, 48) for images of size (160, 160, 96)

# For 'patid' nested output:
# def reset_all_origins(cohort_nii_dir, out_dir):
#     for patient_id in tqdm(os.listdir(cohort_nii_dir)):
#         patient_id_path = os.path.join(cohort_nii_dir, patient_id)
#         for preproc in os.listdir(patient_id_path):
#             preproc_path = os.path.join(patient_id_path, preproc)
#             for nii_name in os.listdir(preproc_path):
#                 nii_path = os.path.join(preproc_path, nii_name)
#                 img = nib.load(nii_path)
#                 img.affine[:3, 3] += OFFSET
                
#                 img.header.set_sform(img.affine)
#                 # img.header.set_qform(img.affine)
#                 nib.save(img, os.path.join(out_dir, nii_name))

# For 'flat' non-nested output:
def reset_all_origins(cohort_nii_dir, out_dir):
    '''
    Set the origins of each image of size (96, 160, 160) to be the center voxel at (48, 80, 80).
    SPM8 sets the default origin in a far corner, so it must be corrected.

    cohort_nii_dir: Directory containing the NII images (as created by convert_all_dicom).
    out_dir: Output directory to write origin-corrected NII files to.
    '''
    if not os.path.exists(cohort_nii_dir):
        raise ValueError(f'Cohort NII directory does not exist: {cohort_nii_dir}')
    if not os.path.exists(out_dir):
        os.mkdir(out_dir)
        print(f'Created NII origin reset directory: "{out_dir}"')
        
    for nii_name in tqdm(os.listdir(cohort_nii_dir)):
        nii_path = os.path.join(cohort_nii_dir, nii_name)
        img = nib.load(nii_path)
        img.affine[:3, 3] += OFFSET
        
        img.header.set_sform(img.affine)
        # img.header.set_qform(img.affine)
        nib.save(img, os.path.join(out_dir, nii_name))
```

```python
# Reset the origins

# reset_all_origins(AD_COHORT_NII_DIR, AD_COHORT_NII_RESET_DIR)
reset_all_origins(MCI_COHORT_NII_DIR, MCI_COHORT_NII_RESET_DIR)
```

```python
def warp_to_template(nii_dir, template_path, out_dir):
    '''
    Warp all images in the directory to the specified template. The images should be in NIFTI1 format, and 
    their origins should overlap with the template within +/- 4cm and +/- 15 degrees
    '''
    if not os.path.exists(nii_dir):
        raise ValueError(f'Cohort NII origin reset directory does not exist: {nii_dir}')
    if not os.path.isfile(template_path):
        raise ValueError(f'Invalid template path: {template_path}')
    if not os.path.exists(out_dir):
        os.mkdir(out_dir)
        print(f'Created NII normalized image directory: {out_dir}')

    eng = matlab.engine.start_matlab()
    eng.cd(os.getcwd())
    for img_name in tqdm(os.listdir(nii_dir)):
        out = eng.normalize_and_write(os.path.join(nii_dir, img_name), template_path, out_dir, nargout=1)
        shutil.move(out['files'][0], out_dir)
        os.remove(out['params'][0])

    eng.quit()


def warp_to_template_parallel(nii_dir, template_path, out_dir, num_engines=12):
    '''
    Parallel version of warp_to_template
    '''
    if not os.path.exists(nii_dir):
        raise ValueError(f'Cohort NII origin reset directory does not exist: {nii_dir}')
    if not os.path.isfile(template_path):
        raise ValueError(f'Invalid template path: {template_path}')
    if not os.path.exists(out_dir):
        os.mkdir(out_dir)
        print(f'Created NII normalized image directory: {out_dir}')
        
    engines = [matlab.engine.start_matlab() for _ in range(num_engines)]
    # engines = [eng.result() for eng in engine_futures]

    img_paths = [os.path.join(nii_dir, img_name) for img_name in os.listdir(nii_dir)]
    args = [(engines[i%num_engines], img_path, template_path, out_dir) for i, img_path in enumerate(img_paths)]
    print('Assembled args')

    results_async = []
    for eng, img_path, _, _ in args:
        out = eng.normalize_and_write(img_path, template_path, out_dir, nargout=1, background=True)
        results_async.append(out)

    print('Began all processes')
    results = [x.result() for x in tqdm(results_async, total=len(results_async))]

    for res in results:
        shutil.move(res['files'][0], out_dir)
        os.remove(res['params'][0])

    for eng in engines:
        eng.quit()

    return results
```
