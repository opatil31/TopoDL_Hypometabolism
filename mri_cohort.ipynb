```python
def write_img_ids(data, date_limit=None):
    '''
    Write the image IDs to a text file separated by commas. Makes it easier to copy-paste the IDs into ADNI ARC builder 
    to download the images.
    '''
    cohorts = data[data['image_date'] < date_limit] if date_limit is not None else data
    
    cn_cohort = cohorts[cohorts['DIAGNOSIS'] == 1.0]['image_id_mri']
    mci_cohort = cohorts[cohorts['DIAGNOSIS'] == 2.0]['image_id_mri']
    ad_cohort = cohorts[cohorts['DIAGNOSIS'] == 3.0]['image_id_mri']
    
    print(f'CN: {len(cn_cohort.unique())}')
    print(f'MCI: {len(mci_cohort.unique())}')
    print(f'AD: {len(ad_cohort.unique())}')
    

    with open('cn_mri_cohort.txt', 'w') as f:
        f.write(','.join(cn_cohort.astype(str)))
        
    with open('mci_mri_cohort.txt', 'w') as f:
        f.write(','.join(mci_cohort.astype(str)))
        
    with open('ad_mri_cohort.txt', 'w') as f:
        f.write(','.join(ad_cohort.astype(str)))
        
        
def write_all_cohorts(data, date_limit=None):
    '''
    Write the fully merged data table to a CSV file. Limit by a maximum date
    '''
    all_cohorts_fname = 'all_cohorts_mri.csv'
    if date_limit is not None:
        data = data[data['image_date'] < date_limit]
        
    data.to_csv(all_cohorts_fname)
```
